```python
# import libraries
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression

from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
 #Load Dataset
df = pd.read_csv("Dataset 2.csv")
df.head()
```

```python
# Q1-Q4
# First five records will be displayed using:
df.head()

# No. of rows and columns
df.shape

# Display column names
df.columns

# Identify numerical and categorical features
numerical = df.select_dtypes(include=['int64', 'float64']).columns
categorical = df.select_dtypes(include = ['object']).columns

print("Numerical features: ", numerical)
print("Categorucal features: ", categorical)
```

```python
#Q5-Q10
#Check missing values
df.isnull().sum()

#Exploratory data analysis
# average age
df["Age"].mean()

#Average watch hours per week
df["WatchHoursPerWeek"].mean()

# Average monthly spending
df["MonthlySpend"].mean()

# users in each subscription category
df["SubscriptionType"].value_counts()

# Percentage of users who renewed subscription
renew_percentage = (df["SubscriptionRenewed"]=="Yes").mean()*100
renew_percentage


```

```python
#Convert Categorical Features into Numerical Form
df_encoded = df.copy()

encoder = LabelEncoder()

for col in ["Gender","SubscriptionType","FavoriteGenre","SubscriptionRenewed"]:
    df_encoded[col] = encoder.fit_transform(df_encoded[col])

df_encoded.head()

# Define Features (X) and Target (y)

X = df_encoded.drop("SubscriptionRenewed", axis=1)
y = df_encoded["SubscriptionRenewed"]

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
#  Train Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

# . Accuracy
prediction = dt.predict(X_test)

accuracy = accuracy_score(y_test, prediction)

print(accuracy)
```

```python
#Q16. Confusion Matrix

confusion_matrix(y_test, prediction)

#Part E: K-Nearest Neighbors (KNN)
#Q17. Train KNN (K = 5)
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

knn_prediction = knn.predict(X_test)

#Q18. Compare Accuracy
knn_accuracy = accuracy_score(y_test, knn_prediction)
print(knn_accuracy)

#Part F: Linear Regression
#Q19. Train Linear Regression Model
X_reg = df_encoded.drop("MonthlySpend", axis=1)
y_reg = df_encoded["MonthlySpend"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.20,
    random_state=42
)

lr = LinearRegression()

lr.fit(X_train_reg, y_train_reg)

# Q20. Predict Monthly Spending for a New User

new_user = X_reg.iloc[[0]]

prediction = lr.predict(new_user)

print(prediction)
```
